# 1. Explicação do dataset
- Housing Prices Dataset é um dataset que fala sobre os preços de casas que fornece diversas informações sobre a casa, como quantidade de quartos e condições da casa, seu principal objetivo é treinar regressão para prever os preços das casas de acordo com as variáveis características das casas.

# 2. Colunas do Dataset


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('train.csv')
print(df.columns)
```

# Descrição das Colunas do Dataset "House Pricing"

O dataset "House Pricing" contém várias colunas que representam características das casas e que podem ser usadas para prever o preço de venda dos imóveis. A seguir estão as principais colunas do dataset, juntamente com seus significados e o tipo de dado (qualitativo ou quantitativo):

### 1. `Id`
- **Significado**: Identificador único da casa no dataset.
- **Tipo**: Qualitativo (categórico).
  
### 2. `MSSubClass`
- **Significado**: Classe de construção do imóvel (tipo de estrutura).
- **Tipo**: Qualitativo (categórico).
  
### 3. `MSZoning`
- **Significado**: Zona de uso do solo. Indica a categoria de zoneamento da propriedade (residencial, comercial, etc.).
- **Tipo**: Qualitativo (categórico).
  
### 4. `LotFrontage`
- **Significado**: Comprimento da frente do terreno em pés.
- **Tipo**: Quantitativo (numérico).
  
### 5. `LotArea`
- **Significado**: Área total do terreno em pés quadrados.
- **Tipo**: Quantitativo (numérico).
  
### 6. `Street`
- **Significado**: Tipo de rua (se é pavimentada ou não).
- **Tipo**: Qualitativo (categórico).

### 7. `Alley`
- **Significado**: Tipo de alley (acesso secundário ao terreno).
- **Tipo**: Qualitativo (categórico).

### 8. `LandContour`
- **Significado**: Contorno do terreno (plano, inclinado, etc.).
- **Tipo**: Qualitativo (categórico).

### 9. `Utilities`
- **Significado**: Tipo de utilidades disponíveis para a casa (água, eletricidade, gás).
- **Tipo**: Qualitativo (categórico).

### 10. `LotConfig`
- **Significado**: Configuração do terreno (se é um terreno interno, esquina, etc.).
- **Tipo**: Qualitativo (categórico).

### 11. `LandSlope`
- **Significado**: Inclinação do terreno (plano, moderadamente inclinado, etc.).
- **Tipo**: Qualitativo (categórico).

### 12. `Neighborhood`
- **Significado**: Bairro onde a casa está localizada.
- **Tipo**: Qualitativo (categórico).

### 13. `Condition1`
- **Significado**: Condição da casa em relação à proximidade de ferrovias, ruas principais, etc.
- **Tipo**: Qualitativo (categórico).

### 14. `Condition2`
- **Significado**: Condição secundária relacionada à casa.
- **Tipo**: Qualitativo (categórico).

### 15. `BldgType`
- **Significado**: Tipo de construção do imóvel (casa, duplex, etc.).
- **Tipo**: Qualitativo (categórico).

### 16. `HouseStyle`
- **Significado**: Estilo arquitetônico da casa (ranch, split-level, etc.).
- **Tipo**: Qualitativo (categórico).

### 17. `OverallQual`
- **Significado**: Qualidade geral da construção (escala de 1 a 10, onde 10 é excelente).
- **Tipo**: Quantitativo (numérico).

### 18. `OverallCond`
- **Significado**: Condição geral da casa (escala de 1 a 10, onde 10 é excelente).
- **Tipo**: Quantitativo (numérico).

### 19. `YearBuilt`
- **Significado**: Ano de construção da casa.
- **Tipo**: Quantitativo (numérico).

### 20. `YearRemodAdd`
- **Significado**: Ano da última remodelação ou adição significativa.
- **Tipo**: Quantitativo (numérico).

### 21. `MasVnrType`
- **Significado**: Tipo de acabamento de parede (pedra, concreto, etc.).
- **Tipo**: Qualitativo (categórico).

### 22. `MasVnrArea`
- **Significado**: Área do acabamento da parede em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 23. `ExterCond`
- **Significado**: Condição do exterior da casa.
- **Tipo**: Qualitativo (categórico).

### 24. `ExterQual`
- **Significado**: Qualidade do material externo da casa.
- **Tipo**: Qualitativo (categórico).

### 25. `Foundation`
- **Significado**: Tipo de fundação (concreto, madeira, etc.).
- **Tipo**: Qualitativo (categórico).

### 26. `BsmtQual`
- **Significado**: Qualidade do porão.
- **Tipo**: Qualitativo (categórico).

### 27. `BsmtCond`
- **Significado**: Condição do porão.
- **Tipo**: Qualitativo (categórico).

### 28. `BsmtExposure`
- **Significado**: Exposição do porão (se ele é parcialmente exposto, subterrâneo, etc.).
- **Tipo**: Qualitativo (categórico).

### 29. `BsmtFinType1`
- **Significado**: Tipo de acabamento do primeiro porão.
- **Tipo**: Qualitativo (categórico).

### 30. `BsmtFinSF1`
- **Significado**: Área do primeiro porão acabado em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 31. `BsmtFinType2`
- **Significado**: Tipo de acabamento do segundo porão (se houver).
- **Tipo**: Qualitativo (categórico).

### 32. `BsmtFinSF2`
- **Significado**: Área do segundo porão acabado em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 33. `BsmtUnfSF`
- **Significado**: Área do porão não acabado em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 34. `TotalBsmtSF`
- **Significado**: Área total do porão em pés quadrados (acabado e não acabado).
- **Tipo**: Quantitativo (numérico).

### 35. `Heating`
- **Significado**: Tipo de aquecimento (gás, elétrico, etc.).
- **Tipo**: Qualitativo (categórico).

### 36. `HeatingQC`
- **Significado**: Qualidade do sistema de aquecimento.
- **Tipo**: Qualitativo (categórico).

### 37. `CentralAir`
- **Significado**: Indica se a casa tem ar condicionado central (Sim ou Não).
- **Tipo**: Qualitativo (categórico).

### 38. `Electrical`
- **Significado**: Tipo de fornecimento elétrico (padrão, tipo de fiação, etc.).
- **Tipo**: Qualitativo (categórico).

### 39. `1stFlrSF`
- **Significado**: Área do primeiro andar em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 40. `2ndFlrSF`
- **Significado**: Área do segundo andar em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 41. `GrLivArea`
- **Significado**: Área da residência acima do solo em pés quadrados.
- **Tipo**: Quantitativo (numérico).

### 42. `BsmtHalfBath`
- **Significado**: Número de banheiros meia-banho no porão.
- **Tipo**: Quantitativo (numérico).

### 43. `FullBath`
- **Significado**: Número de banheiros completos.
- **Tipo**: Quantitativo (numérico).

### 44. `HalfBath`
- **Significado**: Número de banheiros meia-banho (com um lavatório e vaso).
- **Tipo**: Quantitativo (numérico).

### 45. `TotRmsAbvGrd`
- **Significado**: Número total de cômodos acima do nível do solo (excluindo banheiros).
- **Tipo**: Quantitativo (numérico).

### 46. `Fireplaces`
- **Significado**: Número de lareiras na casa.
- **Tipo**: Quantitativo (numérico).

### 47. `GarageType`
- **Significado**: Tipo de garagem (embutida, independente, etc.).
- **Tipo**: Qualitativo (categórico).

### 48. `GarageFinish`
- **Significado**: Acabamento da garagem (se é fechado, sem acabamento, etc.).
- **Tipo**: Qualitativo (categórico).

#4. Medidas Descritivas


```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

medidas_descritivas = pd.DataFrame({
    'Média': df_numeric.mean(),
    'Mínimo': df_numeric.min(),
    'Máximo': df_numeric.max(),
    'Q1': df_numeric.quantile(0.25),
    'Mediana (Q2)': df_numeric.median(),
    'Q3': df_numeric.quantile(0.75)
})

print(medidas_descritivas)
```

#5. Correlação de Pearson

```python
correlation_matrix = df_numeric.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap da Matriz de Correlação')
plt.show()
```

# Heatmap de fácil vizualização

```python
threshold = 0.5
strong_correlation_matrix = correlation_matrix[(correlation_matrix >= threshold) | (correlation_matrix <= -threshold)]

plt.figure(figsize=(12, 8))
sns.heatmap(strong_correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, mask=strong_correlation_matrix.isnull())
plt.title('Heatmap das Correlações Fortes')
plt.show()
```

#6. Visualizações Gráficas

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], kde=True, bins=30)
plt.title('Distribuição do Preço de Venda (SalePrice)')
plt.xlabel('SalePrice')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('Relação entre Área Útil (GrLivArea) e Preço de Venda (SalePrice)')
plt.xlabel('GrLivArea (Área útil)')
plt.ylabel('SalePrice (Preço de Venda)')
plt.show()

plt.figure(figsize=(12, 6))
mean_saleprice_by_quality = df.groupby('OverallQual')['SalePrice'].mean().sort_index()
sns.barplot(x=mean_saleprice_by_quality.index, y=mean_saleprice_by_quality.values, palette="viridis")
plt.title('Preço Médio de Venda por Qualidade Geral (OverallQual)')
plt.xlabel('OverallQual (Qualidade Geral)')
plt.ylabel('Preço Médio de Venda (SalePrice)')
plt.show()
```
